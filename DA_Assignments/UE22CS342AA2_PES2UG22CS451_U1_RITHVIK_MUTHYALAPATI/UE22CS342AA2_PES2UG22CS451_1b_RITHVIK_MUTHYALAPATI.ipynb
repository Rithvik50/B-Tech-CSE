In PCA, the contribution of each feature to a principal component is called the `loading`. Loadings are compared with the absolute value



````r
# Your answer here
# Your answer here
# Load necessary libraries
library(dplyr)
library(ggplot2)

# Load your data
wine_data <- read.csv("/kaggle/input/wine-quality/wine_quality_combined.csv")

# Standardize the data (excluding the 'quality' column)
data_std <- wine_data %>%
  select(-quality) %>%
  scale()

# Perform PCA
pca_result <- prcomp(data_std, center = TRUE, scale. = TRUE)

# Extract standard deviations and compute variance explained
sdev_squared <- pca_result$sdev^2
variance_explained <- sdev_squared / sum(sdev_squared)

# Create a data frame with the proportion of variance explained
variance_df <- data.frame(Principal_Component = 1:length(variance_explained),
                          Proportion_of_Variance = variance_explained)

# Print the data frame
print(variance_df)

# Loadings of the first two principal components
loadings <- pca_result$rotation[, 1:2]

# Print loadings
print(loadings)

# Optional: Visualize loadings (biplot)
if (!requireNamespace("ggbiplot", quietly = TRUE)) {
  install.packages("ggbiplot")
}
library(ggbiplot)

# Visualize loadings (biplot)
ggbiplot(pca_result, choices = 1:2, varname.size = 3)

````



# Dimensionality Reduction and Principal Component Analysis (PCA)

Dimensionality reduction is a process that reduces the number of variables in a dataset while attempting to preserve as much of the original information as possible. Principal Component Analysis (PCA) is a widely-used technique in this context. PCA transforms the data into a new set of coordinates, where the first few principal components account for the largest variance. These components are orthogonal (uncorrelated) and ranked according to the amount of variance they explain in the data.

## Dimensionality Reduction

Techniques like PCA are employed to minimize the number of features in a dataset, while maintaining key patterns. PCA works by converting the data into a different coordinate system, with the principal components ordered by the variance of the data along each axis.

## Proportion of Variance

The `variance_df` provides the proportion of variance explained by each principal component. This measure indicates how much of the total variance in the dataset is captured by each component.

## Scree Plot

A scree plot is useful for deciding how many principal components to keep. The "elbow" of the plot, where the slope begins to level off, often signals the ideal number of components to retain. Components before this point generally capture the majority of the variance, reducing the dimensionality of the dataset while preserving its essential structure.

## Loadings of the First Two Principal Components

The loadings of the first two principal components are available in the `loadings` output. Variables with higher absolute values in these loadings are the most significant contributors to those principal components.

### For the First Principal Component (PC1)

- **Flavanoids**: -0.4237
- **Total_phenols**: -0.3957

**Explanation**: 
- Flavanoids has the highest absolute loading on PC1 (-0.4237), making it the most influential variable for this component.
- Total_phenols follows with a significant loading (-0.3957), also heavily contributing to PC1.

### For the Second Principal Component (PC2)

- **Color_intensity**: -0.5298
- **Hue**: 0.2789

**Explanation**: 
- Color_intensity has the highest absolute loading on PC2 (-0.5298), indicating its substantial impact on this component.
- Hue, with a loading of 0.2789, also plays an important role in PC2.

## Summary

- **PC1**: Strongly influenced by Flavanoids and Total_phenols.
- **PC2**: Primarily affected by Color_intensity and Hue.



*Problem 5* 

Now, for the selected components, find the percentage contribution of each feature to each of the components. (Hint: First find the square of the loadings also known as the cos2 values - find percentage contribution using these values) (2 points)

````r
# Your answer here
# Your answer here
library(tibble)

# Extract the loadings (principal component loadings)
loadings <- pca_result$rotation

# Compute squared loadings (cos2 values)
cos2_values <- loadings^2

# Calculate the percentage contribution for each feature
# Sum of squared loadings for each component
total_cos2 <- apply(cos2_values, 2, sum)

# Compute the percentage contribution
percentage_contribution <- sweep(cos2_values, 2, total_cos2, FUN = "/") * 100

# Display the percentage contribution
print(percentage_contribution)

# Optionally, format and present results
# To display with better formatting
library(dplyr)
percentage_contribution_df <- as.data.frame(percentage_contribution) %>%
  rownames_to_column(var = "Feature") %>%
  mutate(across(starts_with("PC"), ~ round(., 2)))  # rounding for readability

print(percentage_contribution_df)


````
